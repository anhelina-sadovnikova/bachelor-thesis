```python
import pandas as pd
import os
```

```python
csv_path = 'vinbigdata-chest-xray-abnormalities-detection/train.csv'  # Update this with the path to your CSV file
df = pd.read_csv(csv_path)
```

```python
# get the list of image filenames in the specified folder
folder_path = 'segmentation/images_train-kopie'  
image_files = os.listdir(folder_path)
image_files = [file.split('.')[0] for file in image_files]
```

```python
# filter the DataFrame to include only rows where image_id is in image_files
filtered_df = df[df['image_id'].isin(image_files)]
```

```python
# display the filtered DataFrame
filtered_df
```

```python
filtered_df.to_csv('filtered_table_classification.csv', index=False)  # index=False means do not write row numbers
```

```python
# create a new DataFrame to hold the unique image IDs and diagnosis flags
unique_images_df = pd.DataFrame(filtered_df['image_id'].unique(), columns=['image_id'])
```

```python
diagnoses = [
    'Atelectasis', 'Calcification', 'Cardiomegaly', 'Consolidation', 'ILD', 'Infiltration',
    'Lung Opacity', 'Nodule', 'Other Lesion', 'Pleural Effusion', 'Pleural Thickening',
    'Pneumothorax', 'Pulmonary fibrosis', 'Aortic enlargement'
]
```

```python
# find unique class names and create a DataFrame with these diagnosis columns initialized to 0
unique_class_names = filtered_df['class_name'].unique()
columns = ['image_id'] + list(unique_class_names)
unique_images_df = pd.DataFrame(columns=columns)
unique_images_df['image_id'] = filtered_df['image_id'].unique()

# initialize all diagnosis columns to 0
for class_name in unique_class_names:
    unique_images_df[class_name] = 0

# set diagnosis flags
for _, row in filtered_df.iterrows():
    image_id = row['image_id']
    class_name = row['class_name']
    unique_images_df.loc[unique_images_df['image_id'] == image_id, class_name] = 1

# display the resultant DataFrame
unique_images_df
```

```python
unique_images_df.to_csv('unique_images_df.csv', index=False)  # index=False means do not write row numbers
```
